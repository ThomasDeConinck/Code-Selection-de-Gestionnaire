```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Charger les données depuis la WorkSheet Excel
df_EAFE = pd.read_excel('/Users/marievegendron/Documents/GitHub/Code-Selection-de-Gestionnaire/EAFE.xlsx', sheet_name='VaR by country')
df_EAFE
```

```python
# Charger les données depuis la WorkSheet Excel
df_EM = pd.read_excel('/Users/marievegendron/Documents/GitHub/Code-Selection-de-Gestionnaire/EM.xlsx', sheet_name='VaR by country')
df_EM
```

**Graphique - EAFE**

```python
pays = df_EAFE['Country'].tolist()
var_millions = df_EAFE['VaR (MC)'].astype(float).tolist()
poids = df_EAFE['Weight (%)'].astype(float).tolist()  

# Création du graphique
fig, ax = plt.subplots(figsize=(10, 6)) 
nuage = ax.scatter(var_millions, poids, alpha=0.7, edgecolors='black')

for i, pays in enumerate(pays):
    if pays == "Spain":
        ax.annotate(pays, (var_millions[i], poids[i]), fontsize=9, ha='center', va='bottom', color='black', style='italic')
    elif pays == "Switzerland":
        ax.annotate(pays, (var_millions[i], poids[i]), fontsize=9, ha='left', va='bottom', color='black', style='italic')
    else:
        ax.annotate(pays, (var_millions[i], poids[i]), fontsize=9, ha='center', va='bottom', color='black', style='italic')
        
ax.set_xlabel('VaR (MC)', fontsize=12, weight='bold')
ax.set_ylabel('Poids du portefeuille (%)', fontsize=12, weight='bold')
ax.set_title('Analyse du risque de portefeuille par pays', fontsize=14, weight='bold')

ax.set_xlim([0, max(var_millions)*1.1])
ax.set_ylim([0, max(poids)*1.1])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

**Graphique - EM**

```python
rows_to_keep = df_EM[df_EM['Weight (%)'] >= 1.5]
var_millions = rows_to_keep['VaR (MC)'].astype(float).tolist()
poids = rows_to_keep['Weight (%)'].astype(float).tolist()
pays = rows_to_keep['Country'].astype(str).tolist()

# Création du graphique
fig, ax = plt.subplots(figsize=(10, 6)) 
nuage = ax.scatter(var_millions, poids, alpha=0.7, edgecolors='black')

for i, pays in enumerate(pays):
    if pays == "China":
        ax.annotate(pays, (var_millions[i], poids[i]), fontsize=9, ha='center', va='bottom', color='black', style='italic')
    elif pays == "Mexico":
        ax.annotate(pays, (var_millions[i], poids[i]), fontsize=9, ha='left', va='bottom', color='black', style='italic')
    else:
        ax.annotate(pays, (var_millions[i], poids[i]), fontsize=9, ha='center', va='bottom', color='black', style='italic')

        
ax.set_xlabel('VaR (MC)', fontsize=12, weight='bold')
ax.set_ylabel('Poids du portefeuille (%)', fontsize=12, weight='bold')
ax.set_title('Analyse du risque de portefeuille par pays', fontsize=14, weight='bold')

ax.set_xlim([0, max(var_millions)*1.1])
ax.set_ylim([0, max(poids)*1.1])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
